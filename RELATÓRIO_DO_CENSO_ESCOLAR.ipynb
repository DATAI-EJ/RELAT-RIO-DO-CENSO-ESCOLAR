```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly import graph_objects as go
from tabulate import tabulate
from prettytable import PrettyTable
import plotly.figure_factory as ff

sns.set(style="whitegrid")

df = pd.read_csv('/content/arquivo_filtrado.csv',sep=';')

print("Primeiras linhas do DataFrame:")
df.head()
```

```python
print("\nInformações gerais do DataFrame:")
df.info()
```

```python
print(df.columns)
```

```python
acessibilidade_cols = [
    'IN_ACESSIBILIDADE_CORRIMAO',
    'IN_ACESSIBILIDADE_ELEVADOR',
    'IN_ACESSIBILIDADE_PISOS_TATEIS',
    'IN_ACESSIBILIDADE_VAO_LIVRE',
    'IN_ACESSIBILIDADE_RAMPAS',
    'IN_ACESSIBILIDADE_SINAL_SONORO',
    'IN_ACESSIBILIDADE_SINAL_TATIL',
    'IN_ACESSIBILIDADE_SINAL_VISUAL'
]

df[acessibilidade_cols] = df[acessibilidade_cols].replace({1: 'Sim', 0: 'Não'})
df['IN_INTERNET'] = df['IN_INTERNET'].replace({1: 'Sim', 0: 'Não'})

acessibilidade_stats = df[acessibilidade_cols + ['IN_INTERNET']].apply(lambda x: x.value_counts()).fillna(0)
acessibilidade_percent = df[acessibilidade_cols + ['IN_INTERNET']].apply(lambda x: x.value_counts(normalize=True) * 100).fillna(0).round(2)

def print_pretty_table(df, title):
    table = PrettyTable()
    table.title = title
    table.field_names = ["Recurso"] + list(df.columns)
    for index, row in df.iterrows():
        table.add_row([index] + list(row))
    print(table)

print_pretty_table(acessibilidade_stats, "Estatísticas de Acessibilidade e Internet (Quantidades de Sim/Não)")

print_pretty_table(acessibilidade_percent, "Percentuais de Acessibilidade e Internet (Sim/Não)")
```

```python
nomes_formatados = {
    'IN_ACESSIBILIDADE_CORRIMAO': 'Corrimão',
    'IN_ACESSIBILIDADE_ELEVADOR': 'Elevador',
    'IN_ACESSIBILIDADE_PISOS_TATEIS': 'Pisos Táteis',
    'IN_ACESSIBILIDADE_VAO_LIVRE': 'Vão Livre',
    'IN_ACESSIBILIDADE_RAMPAS': 'Rampas',
    'IN_ACESSIBILIDADE_SINAL_SONORO': 'Sinal Sonoro',
    'IN_ACESSIBILIDADE_SINAL_TATIL': 'Sinal Tátil',
    'IN_ACESSIBILIDADE_SINAL_VISUAL': 'Sinal Visual',
    'IN_INTERNET': 'Internet'
}
```

```python
situacao_labels = {
    1: "Em Atividade",
    2: "Paralisada",
    3: "Extinta (Ano do Censo)",
    4: "Extinta em Anos Anteriores"
}

df['TP_SITUACAO_FUNCIONAMENTO_DESC'] = df['TP_SITUACAO_FUNCIONAMENTO'].map(situacao_labels)
situacao_counts = df['TP_SITUACAO_FUNCIONAMENTO_DESC'].value_counts()

fig = px.bar(
    situacao_counts,
    x=situacao_counts.index,
    y=situacao_counts.values,
    labels={"x": "Situação de Funcionamento", "y": "Quantidade de Escolas"},
    title="Distribuição das Situações de Funcionamento das Escolas",
    text=situacao_counts.values
)
fig.update_layout(
    width=800,
    height=500
)
fig.update_traces(textposition="outside", marker_color="#1f77b4")
fig.update_layout(xaxis_title="Situação de Funcionamento", yaxis_title="Quantidade de Escolas", xaxis_tickangle=45)
fig.show()
```

```python
df['TP_SITUACAO_FUNCIONAMENTO'] = df['TP_SITUACAO_FUNCIONAMENTO'].replace({
    1: 'Em Atividade',
    2: 'Paralisada',
    3: 'Extinta (Ano do Censo)',
    4: 'Extinta em Anos Anteriores'
})

internet_por_situacao = df.groupby(['TP_SITUACAO_FUNCIONAMENTO', 'IN_INTERNET']).size().reset_index(name='Quantidade')
for situacao in internet_por_situacao['TP_SITUACAO_FUNCIONAMENTO'].unique():
    dados_situacao = internet_por_situacao[internet_por_situacao['TP_SITUACAO_FUNCIONAMENTO'] == situacao]

    fig = px.pie(
        dados_situacao,
        names='IN_INTERNET',
        values='Quantidade',
        title=f"Distribuição de Internet",
        color_discrete_map={'Não': '#d62728', 'Sim': '#2ca02c'}
    )
    fig.update_layout(width=500, height=500)
    fig.show()
```

```python
df_formatado = df.rename(columns=nomes_formatados)
contagem_acessibilidade = df_formatado[nomes_formatados.values()].apply(lambda col: col.value_counts()).fillna(0).T
contagem_acessibilidade_percentual = contagem_acessibilidade.div(contagem_acessibilidade.sum(axis=1), axis=0) * 100
fig = go.Figure()

for recurso in contagem_acessibilidade_percentual.columns:
    fig.add_trace(
        go.Bar(
            name=recurso,
            x=contagem_acessibilidade_percentual.index,
            y=contagem_acessibilidade_percentual[recurso],
            text=contagem_acessibilidade_percentual[recurso].round(1).astype(str) + '%',
            textposition='auto',
            marker_color={'Sim': '#2ca02c', 'Não': '#d62728'}[recurso]
        )
    )

fig.update_layout(
    barmode='stack',
    title='Distribuição de Recursos de Acessibilidade nas Escolas (Porcentagem)',
    xaxis_title='Recursos de Acessibilidade',
    yaxis_title='Porcentagem (%)',
    xaxis=dict(tickangle=45, tickmode='linear'),
    legend_title='Disponibilidade',
    template='plotly_white',
    title_font_size=20,
    font=dict(size=14),
    height=600,
    width=1000
)

fig.show()
```

```python
df['IN_INTERNET'] = df['IN_INTERNET'].replace({1: 'Sim', 0: 'Não'})

internet_por_estado = df.groupby(['NO_REGIAO', 'IN_INTERNET']).size().unstack(fill_value=0)
internet_por_estado_percent = internet_por_estado.div(internet_por_estado.sum(axis=1), axis=0) * 100

fig = go.Figure()

for status in ['Sim', 'Não']:
    fig.add_trace(go.Bar(
        x=internet_por_estado_percent.index,
        y=internet_por_estado_percent[status],
        name=status,
        text=internet_por_estado_percent[status].apply(lambda x: f"{x:.1f}%"),
        textposition='inside',
        marker_color={'Sim': '#2ca02c', 'Não': '#d62728'}[status]
    ))

fig.update_layout(
    title=f"Proporção de Escolas com Internet por Região (%)",
    xaxis_title="Região",
    yaxis_title="Porcentagem de Escolas",
    barmode='stack',
    legend=dict(title='Internet'),
    template="plotly_white",
    width=900,
    height=600
)

fig.show()
```

```python
nomes_formatados_acc = {
    'IN_ACESSIBILIDADE_CORRIMAO': 'Corrimão',
    'IN_ACESSIBILIDADE_ELEVADOR': 'Elevador',
    'IN_ACESSIBILIDADE_PISOS_TATEIS': 'Pisos Táteis',
    'IN_ACESSIBILIDADE_VAO_LIVRE': 'Vão Livre',
    'IN_ACESSIBILIDADE_RAMPAS': 'Rampas',
    'IN_ACESSIBILIDADE_SINAL_SONORO': 'Sinal Sonoro',
    'IN_ACESSIBILIDADE_SINAL_TATIL': 'Sinal Tátil',
    'IN_ACESSIBILIDADE_SINAL_VISUAL': 'Sinal Visual',
}


df_formatado = df.rename(columns=nomes_formatados_acc)
acessibilidade_por_regiao = df_formatado.groupby('NO_REGIAO')[list(nomes_formatados_acc.values())] \
    .apply(lambda x: x.replace({'Sim': 1, 'Não': 0}).astype(float).mean())

dados_radar = acessibilidade_por_regiao.reset_index().melt(
    id_vars='NO_REGIAO',
    var_name='Recurso',
    value_name='Porcentagem'
)

fig = px.line_polar(
    dados_radar,
    r='Porcentagem',
    theta='Recurso',
    color='NO_REGIAO',
    line_close=True,
    template='plotly_white',
    title='Acessibilidade por Região (Radar Chart)'
)

fig.update_traces(
    line=dict(width=3),
    marker=dict(size=8)
)

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 1],
            tickfont=dict(size=12),
            tickcolor='gray'
        ),
        angularaxis=dict(
            tickfont=dict(size=12),
        )
    ),
    title_font_size=22,
    font=dict(size=14),
    height=600,
    width=500,
    legend=dict(
        title='Regiões',
        font=dict(size=14),
        orientation='h',
        yanchor='bottom',
        y=-0.2,
    )
)

fig.show()
```

```python
situacao_labels = {
    1: "Em Atividade",
    2: "Paralisada",
    3: "Extinta (Ano do Censo)",
    4: "Extinta em Anos Anteriores"
}

df_formatado = df.rename(columns=nomes_formatados)

df['TP_SITUACAO_FUNCIONAMENTO_DESC'] = df['TP_SITUACAO_FUNCIONAMENTO'].map(situacao_labels)

def contar_acessibilidades(grupo):
    estatisticas = grupo[list(nomes_formatados.values())].apply(pd.Series.value_counts, axis=0).fillna(0)
    estatisticas = estatisticas.rename(index={1: '1 (Ativo)', 0: '0 (Inativo)'})
    total = estatisticas.sum(axis=0)
    porcentagens = estatisticas.div(total) * 100
    resultado = pd.concat([estatisticas, porcentagens], keys=['Contagem', 'Porcentagem'])

    return resultado
df_filtrado = df_formatado[df['TP_SITUACAO_FUNCIONAMENTO_DESC'] == "Em Atividade"]

acessibilidade_por_regiao = df_filtrado.groupby('NO_REGIAO').apply(contar_acessibilidades)

print(tabulate(acessibilidade_por_regiao, headers='keys', tablefmt='psql'))
```
